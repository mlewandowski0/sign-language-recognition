```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv('wandb_export_MHI-old-algo.csv')
```

```python
dataframe
```

```python
#labels unique class  so add network type and size
# Extract the necessary columns
architecture = dataframe['CNN Architecture']
val_accuracy = dataframe['epoch/val_categorical_accuracy (Max)']
```

## Architecture effect on Validation Accuracy

```python
#Plotting
# Plotting
plt.figure(figsize=(10, 6))
for arch in architecture.unique():
    # Filter the DataFrame for each architecture and plot their accuracies
    subset = dataframe[dataframe['CNN Architecture'] == arch]['epoch/val_categorical_accuracy (Max)']
    plt.hist(subset, alpha=0.5, label=arch)

plt.xlabel('Validation Accuracy')
plt.ylabel('Frequency')
plt.title('Histogram of Validation Accuracy for different CNN Architectures')
plt.legend(title='CNN Architecture')
plt.show()
```

```python
# Assuming 'dataframe' is already defined and has the necessary data
# Group by 'CNN Architecture' and get the maximum 'epoch/val_categorical_accuracy (Max)' for each
grouped_data = dataframe.groupby('CNN Architecture')['epoch/val_categorical_accuracy (Max)'].max()

# Plotting
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', alpha=0.75)

plt.xlabel('CNN Architecture')
plt.ylabel('Max Validation Accuracy')
plt.title('Max Validation Accuracy for Different CNN Architectures')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout for better fit of x-axis labels
plt.show()
```

```python

# Load data
dataframe = pd.read_csv('wandb_export_MHI-old-algo.csv')

# Pre-calculate the maximum validation accuracy for each architecture
max_accuracy_per_architecture = dataframe.groupby('CNN Architecture', as_index=False)['epoch/val_categorical_accuracy (Max)'].max()

# Plotting
plt.figure(figsize=(10, 8))  # Adjust width to 10 for a narrower plot
barplot = sns.barplot(
    data=max_accuracy_per_architecture,
    x='CNN Architecture',
    y='epoch/val_categorical_accuracy (Max)',
    edgecolor='black',
    palette='viridis'
)

# Adding annotations for each bar with increased font size
for p in barplot.patches:
    height = p.get_height()
    barplot.annotate(
        format(height, '.2f'), 
        (p.get_x() + p.get_width() / 2., height), 
        ha='center', va='center', 
        xytext=(0, 9), 
        textcoords='offset points',
        fontsize=12,  # Increased font size for annotations
        color='black'
    )

plt.xlabel('2DCNN Architectures', fontsize=15)  # Increased font size for x label
plt.ylabel('Max Validation Accuracy', fontsize=15)  # Increased font size for y label
plt.title('Maximum Validation Accuracy for Different CNN Architectures', fontsize=22)  # Increased font size for title
plt.xticks(rotation=0, ha='center', fontsize=12)  # Rotate and increase font size of x-axis labels
plt.tight_layout()  # Adjust layout for better fit of x-axis labels
plt.grid(axis='y')  # Add a grid for the y axis
plt.savefig('1-chan_MHI_max_validation_accuracy.png', dpi=300)  # Save with high resolution
plt.show()


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
dataframe = pd.read_csv('wandb_export_MHI-old-algo.csv')

# Create a unique identifier for each run within each architecture
dataframe['Run Identifier'] = dataframe.groupby('CNN Architecture').cumcount() + 1

# Plotting
plt.figure(figsize=(14, 8))
barplot = sns.barplot(
    data=dataframe,
    x='CNN Architecture',
    y='epoch/val_categorical_accuracy (Max)',
    hue='Run Identifier',  # This will create a separate bar for each run within the architecture
    dodge=True,  # Ensures bars are side-by-side
    palette='viridis',  # Adjust color palette for visibility
    edgecolor='black',
    linewidth=0.5  # Line width for edges
)

plt.xlabel('CNN Architecture')
plt.ylabel('Validation Accuracy')
plt.title('Validation Accuracies for Each Run by CNN Architecture')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Run ID', bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside of plot
plt.tight_layout()  # Adjust layout for better fit of x-axis labels
plt.grid(axis='y')  # Add a grid for the y axis
plt.show()
```

## Data Augmentation Effects

```python

# Load the CSV file into a pandas DataFrame
dataframe = pd.read_csv('wandb_export_MHI-old-algo.csv')

# Create a unique identifier for each run within each data augmentation category
# Convert the run identifier to string to avoid errors during plotting
dataframe['Run Identifier'] = (dataframe.groupby('Data Augmentation?').cumcount() + 1).astype(str)

# Plotting
plt.figure(figsize=(10, 6))
barplot = sns.barplot(
    data=dataframe,
    x='Data Augmentation?',
    y='epoch/val_categorical_accuracy (Max)',
    hue='Run Identifier',  # This will create a separate bar for each run within the data augmentation category
    dodge=True,  # Ensures bars are side-by-side
    palette='viridis',  # Adjust color palette for visibility
    edgecolor='black',
    linewidth=0.5  # Line width for edges
)

plt.xlabel('Data Augmentation', fontsize=12)
plt.ylabel('Validation Accuracy', fontsize=12)
plt.title('Effect of Data Augmentation on Validation Accuracy', fontsize=15)
plt.xticks(rotation=0, ha='center')  # Adjust rotation and alignment for better readability
plt.legend(title='Run ID', bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside of plot to avoid overlap
plt.tight_layout()  # Adjust layout for better fit of x-axis labels
plt.grid(axis='y')  # Add a grid for the y axis
plt.savefig('MHI-1-chan-data-aug.png', dpi = 300)  # Save with high resolution
plt.show()

```
