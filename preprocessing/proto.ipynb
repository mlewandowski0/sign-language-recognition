```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0, 1, 500)

v1, v2, v3 = 1/3, 2/3, 1.

s1 = np.clip((1 - 9*(v1 - x)**2), 0, 1)
s2 = np.clip((1.0 - 9*(v2 - x)**2), 0, 1)
s3 = np.clip((1.0 - 4*(v3 - x)**2), 0, 1)


plt.plot(x, s1, label="s1", color="red")
plt.plot(x, s2, label="s2", color="blue")
plt.plot(x, s3, label="s3", color="green")
plt.legend()
plt.show()
```

```python


def pseudocolor_generator(X, v1, v2, v3, v1_node, v2_node, v3_node, pow=1):
    s1 = np.clip(1 - ((1./((v1_node - v1)**2)) * (v1 - X)**2  )**pow, 0, 1)
    s2 = np.clip(1 - ((1./((v2_node - v2)**2)) * (v2 - X)**2  )**pow, 0, 1)
    s3 = np.clip(1 - ((1./((v3_node - v3)**2)) * (v3 - X)**2  )**pow, 0, 1)

    return s1, s2, s3

def pseudocol_plot(x, v1, v2, v3, v1_node, v2_node, v3_node, pow=1):
    s1, s2, s3 = pseudocolor_generator(x, v1, v2, v3, v1_node, v2_node, v3_node, pow=pow)
    plt.title(f"v1={round(v1,2)},v2={round(v2,2)},v3={round(v3,2)},v1_node={round(v1_node,2)}, v2_node={round(v2_node,2)}, v3_node={round(v3_node,2)}, pow={round(pow, 2)}")
    plt.plot(x, s1, label="s1", color="blue")
    plt.plot(x, s2, label="s2", color="green")
    plt.plot(x, s3, label="s3", color="red")
    plt.legend()

def pseudocol_plot(x, v1, v2, v3, v1_node, v2_node, v3_node, pow=1):
    s1, s2, s3 = pseudocolor_generator(x, v1, v2, v3, v1_node, v2_node, v3_node, pow=pow)
    plt.title(f"v1={round(v1,2)},v2={round(v2,2)},v3={round(v3,2)},v1_node={round(v1_node,2)}, v2_node={round(v2_node,2)}, v3_node={round(v3_node,2)}, pow={round(pow, 2)}", fontdict={"fontsize" : 12})
    plt.plot(x, s1, color="blue", label="blue channel")
    plt.plot(x, s2, color="green", label="green channel")
    plt.plot(x, s3, color="red", label="red channel")
    plt.xlabel("MHI Intensity", fontdict={"fontsize" : 14})
    plt.ylabel("Per channel Intenstity", fontdict={"fontsize" : 14})
    plt.legend()
    plt.grid()
    
x = np.linspace(0, 1, 500)
pseudocol_plot(x, 1/3, 2/3, 1, 0, 1, 0.5, pow=1)
plt.savefig("pseudcolor_vis1.png")
plt.show()

pseudocol_plot(x, 1/3, 2/3, 1, 0, 1, 0.5, pow=0.5)
plt.savefig("pseudcolor_vis2.png")
plt.show()

pseudocol_plot(x, 1/3, 2/3, 1, 0, 1, 0.5, pow=0.25)
plt.savefig("pseudcolor_vis3.png")
plt.show()

pseudocol_plot(x, 1/3, 2/3, 1, 0, 1, 0.5, pow=2)
plt.savefig("pseudcolor_vis4.png")
plt.show()
```
